# 5.2 Decision Boundary -- PyTorch ðŸ”¥

**Exercise:** Use PyTorch to find a decision boundary for point in the make_moons.

**Hint:** Use the next function to plt the decision boundary,

```python
# Visualizar el dataset
plt.figure(figsize=(12, 8))
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train, s=40, cmap=plt.cm.Spectral)
plt.show()
```

```python
#2. Definir el modelo de red neuronal en PyTorch
#Voy a definir una red neuronal simple de dos capas en PyTorch.
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.autograd import Variable

# Definir el modelo de red neuronal
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.layer1 = nn.Linear(2, 10)
        self.layer2 = nn.Linear(10, 1)

    def forward(self, x):
        x = torch.relu(self.layer1(x))
        x = torch.sigmoid(self.layer2(x))
        return x
#Se define el modelo se tienen inicialmente 2 capas 10 salidas, luego despues toma 10 capas y una salida aplicado a las dos  funciones de activaciÃ²n
```

```python
#3. Convertir los datos a tensores de PyTorch
```

```python
# Convertir los datos a tensores
x_train_tensor = torch.tensor(x_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)

#Se realiza para poder ubicar los tensores en una sola dimension.
```

```python
#4. Entrenar el modelo
#Se utiliza el parametro
#Utilizaremos un optimizador como Adam y la funciÃ³n de pÃ©rdida BCELoss (ya que se trata de una tarea de clasificaciÃ³n binaria).
```

```python
# Aplicar el modelo, la funciÃ³n de pÃ©rdida y el optimizador
model = SimpleNN()
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Criterion : Funcion de Perdida con entriopia para RNN

# Entrenamiento del modelo
epochs = 1000
for epoch in range(epochs):
    model.train()
    # Forward pass
    y_pred = model(x_train_tensor)
    loss = criterion(y_pred, y_train_tensor)

    # Backward pass y optimizaciÃ³n
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Imprimir el progreso cada 100 Ã©pocas
    if (epoch+1) % 100 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')
```

```python
#5. FunciÃ³n para trazar el lÃ­mite de decisiÃ³n
#La funciÃ³n plot_decision_boundary que definimos antes.
#  Esta funciÃ³n trazarÃ¡ el lÃ­mite de decisiÃ³n basado en el modelo entrenado.
```

```python
import numpy as np
from sklearn.metrics import roc_auc_score
import torch
from torch.autograd import Variable

def plot_decision_boundary(dataset, labels, model, steps=1000, color_map='Paired'):
    color_map = plt.get_cmap(color_map)
    # Definir los lÃ­mites de la regiÃ³n de interÃ©s
    xmin, xmax = dataset[:, 0].min() - 1, dataset[:, 0].max() + 1
    ymin, ymax = dataset[:, 1].min() - 1, dataset[:, 1].max() + 1
    x_span = np.linspace(xmin, xmax, steps)
    y_span = np.linspace(ymin, ymax, steps)
    xx, yy = np.meshgrid(x_span, y_span)

    # Hacer predicciones en la regiÃ³n de interÃ©s
    model.eval()
    grid = np.c_[xx.ravel(), yy.ravel()]
    labels_predicted = model(Variable(torch.tensor(grid, dtype=torch.float32)))

    # Procesar las predicciones
    labels_predicted = (labels_predicted > 0.5).float().detach().numpy().reshape(xx.shape)

    # Dibujar el lÃ­mite de decisiÃ³n
    fig, ax = plt.subplots()
    ax.contourf(xx, yy, labels_predicted, cmap=color_map, alpha=0.5)

    # Dibujar los puntos de entrenamiento
    ax.scatter(dataset[:, 0], dataset[:, 1], c=labels.reshape(labels.size), cmap=color_map, lw=0)
    plt.show()

    # Calcular y mostrar el AUC
    with torch.no_grad():
        # Predicciones en formato de probabilidad para el dataset original
        y_pred_prob = model(Variable(torch.tensor(dataset, dtype=torch.float32)))
        auc = roc_auc_score(labels, y_pred_prob.numpy())
        print(f'AUC: {auc:.4f}')

# Trazar el lÃ­mite de decisiÃ³n y calcular el AUC
plot_decision_boundary(x_train, y_train, model)
```
